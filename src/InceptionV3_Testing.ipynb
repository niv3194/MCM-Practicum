```python
from keras.models import load_model
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import numpy as np
import cv2
from scipy.misc import imresize
```

```python
names = {'parade': 0, 'handshaking': 1, 'demonstration': 2, 'riot': 3, 'cheering': 4, 'shoppers': 5, 'soccer': 6}
```

```python
import pandas as pd
```

```python
model = load_model('C:/Users/Nivedha/Documents/css/7_modelInceptionV31.h5')
```

```python
def get_name(names, location):
    """
    Reads in the appropriate dictionary of classes and the location of the class we want
    :param names: dictionary of classes and integer labels
    :param location: integer label of event
    :return: the name of the event that lines up with the passes location
    """
    for name in names:
        if names[name] == location:
            return name
```

```python
import pandas as pd
import os
dim = 299  # image height and width dimensions
default_path = 'C:/Users/Nivedha/Desktop/test_data/'
#print('Please enter the full name of the image to predict.\nNote: must be in directory predict/to_predict/\n')
five = []
array = []
for x in os.listdir(default_path):
    #loc = 0
    if x.endswith(".jpg"):
        file=x
            
        
        # read the image and resize it to a fixed-size
        image = cv2.imread(default_path + '/' + file)
        print(file)
        #array = []   
        array.append(file)# grab user input for the path of image to predict
        #df=pd.DataFrame(array)
        #print(df)
        #files = pd.DataFrame(df,columns=['image']).to_csv('C:/Users/Nivedha/Documents/Memorability/image.csv')
        if file == 'exit':
            exit(0)
        img = load_img(default_path + file)
        x = imresize(img_to_array(img), (dim, dim, 3))
        x = x.reshape((1,) + x.shape)  # get the proper # of dimensions - only predict 1 image at a time

        predict_d = ImageDataGenerator(rescale=1. / 255)  # regularize
    # preprocess the image in an infinite loop
        for batch in predict_d.flow(x, batch_size=1):
            x = batch  # this runs in an infinite loop - need to stop it
            break

        prediction = model.predict(x, 1)
        #print(prediction)
        sum = np.sum(prediction)
        percentages = [] 
        #loc = 0
        event=''# standard python list to contain the percentages

    # to calculate the percentage take each independent probability and divide it by the sum of all
        for prob in np.nditer(prediction):
            percentages.append((prob / sum) * 100)
            
        #print(pd.DataFrame(percentages))
        print(percentages)
        
        loc = 0
        for percent in percentages:
            if(percent>45):
                pred = (percent, get_name(names, loc))
                print(pred)
                five.append(pred)
                #predictions = pd.DataFrame(pred,columns=['confidence']).to_csv('C:/Users/Nivedha/Documents/Memorability/check.csv')
            loc+= 1
```
